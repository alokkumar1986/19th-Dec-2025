```python
#write a program to find maximum of three
def findMaxNumber(x, y, z):
  if x>y and x >z:
    print("{} is maximum".format(x))
  elif y>z:
    print("{} is maximum".format(y))
  else:
    print("{} is maximum".format(z))

num1 = int(input('Enter first number : '))
num2 = int(input('Enter second number : '))
num3 = int(input('Enter third number : '))
findMaxNumber(num1, num2, num3)
print('End')
```

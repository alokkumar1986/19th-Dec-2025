```python
l = ['apple',  'orange', 'papaya', 'banana']

print(l[1:3]) # ['orange', 'papaya']

print(l[:3]) # ['apple',  'orange', 'papaya']

print(l[1:]) # ['orange', 'papaya', 'banana']

print(l[-3:-1])

print(l[:])

print(l[::-1])
```

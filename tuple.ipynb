```python
a = (10, 20, 30, 40, 50)

print(a[1:3]) #(20, 30)

print(a[:3]) # (10, 20, 30)

print(a[1:]) # (20, 30, 40, 50)

print(a[-3:-1]) # (30, 40)

print(a[:]) # (10, 20, 30, 40, 50)

print(a[::-1]) # (50, 40, 30, 20, 10)

```

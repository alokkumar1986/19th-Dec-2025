```python
A = {"Apple", "Banana", "Orange"}
B = {1, 3, 45, 7, 66}

C = A.union(B)
print(C)
```

```python
A = {"Apple", "Banana", "Orange"}
B = {1, 3, 45, 7, 66}

C = A.intersection(B)
print(C)
```

```python
A = {"Apple", "Banana", "Orange"}
B = {1, 3, 45, 7, 66}

C = A.difference(B)
print(C)
```

```python
A = {"Apple", "Banana", "Orange"}
B = {1, 3, 45, 7, 66}

C = A.symmetric_difference(B)
print(C)
```

```python
A = {"Apple", "Banana", "Orange"}
B = {1, 3, 45, 7, 66}
C = { 1.2, 3.7, 1.3, 1.5}

D = A | B | C

#D = A.union(B, C)
print(D)
```

```python
A = {"Apple", "Banana", "Orange"}
B = {1, 3, 45, 7, 66}
C = { 1.2, 3.7, 1.3, 1.5}

#D = A & B & C

D = A.intersection(B, C)
print(D)
```

```python
A = {"Apple", "Banana", "Orange"}
B = {1, 3, 45, 7, 66}
C = { 1.2, 3.7, 1.3, 1.5}

#D = A - B - C

D = A.difference(B, C)
print(D)
```

```python
A = {"Apple", "Banana", "Orange"}
B = {1, 3, 45, 7, 66}
C = { 1.2, 3.7, 1.3, 1.5}

#D = A ^ B ^ C

D = A.symmetric_difference(B, C)
print(D)
```

```python
#Create a dictionary called book with keys: "title", "author", "year", and "pages".
#Fill it with info about your favorite book. Then print each value using both bracket notation and .get()

book = {
  "book1" :  {
    "title" : "Python Programming",
    "author" : "S Sharma",
    "year" : 2023,
    "pages" : 500
  },
  "book2" : {
    "title" : "Python For Beginers",
    "author" : "P Sharma",
    "year" : 2022,
    "pages" : 400
  }
}

print(book['book1']['title'])
print(book.get('book1').get('title'))

print(book['book1']['author'])
print(book.get('book1').get('author'))

print(book['book1']['year'])
print(book.get('book1').get('year'))

print(book['book1']['pages'])
print(book.get('book1').get('pages'))

#=============================================================

print(book['book2']['title'])
print(book.get('book2').get('title'))

print(book['book2']['author'])
print(book.get('book2').get('author'))

print(book['book2']['year'])
print(book.get('book2').get('year'))

print(book['book2']['pages'])
print(book.get('book2').get('pages'))


```

```python
# Starting with menu = {"burger": 8.99, "fries": 3.49},
# add "soda" at 1.99,
# change "burger" to 9.49,
# remove "fries", and
# print the final menu with a loop


menu = {"burger": 8.99, "fries": 3.49}
menu['soda'] = 1.99
menu['burger'] = 9.49
menu.pop('fries')
for keys, val in menu.items():
    print(keys, val)
```

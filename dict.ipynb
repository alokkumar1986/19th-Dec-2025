```python
students ={

   0 : {
      "name" : "ABCD",
      "age" : 23,
      "course" : ["Python", "MySQL", "POWER BI"]
   },
   1 : {
      "name" : "PQRS",
      "age" : 25,
      "course" : ["Python", "MySQL"]
   }

}
print(students)
print(students[0])
print(students[0].get('age'))
print(students[0]['age'])

print(students[1].get("name"))
print(students[1]['name'])

print(students[1]['course'][1])

print(students[0]['course'][2])
```
